## Построение матрицы корреляции

```python
columns = ["teaching", "international", "research", "citations", "income", "total_score", "num_students"]
data_array = data.as_matrix(columns = columns)
data_array_ma = np.ma.masked_invalid(data_array)
corr_coef = np.ma.corrcoef(data_array_ma, rowvar = 0)
corr_coef_df = pd.DataFrame(corr_coef, columns= columns, index=columns)
print "Pearson correlation matrix:"
corr_coef_df
```

## Оценка значимости коэффициентов корреляции

### t-критерий Стьюдента

```python
def significance_test(r, n, t_critical):
    t_test = abs(r)*sqrt(n-2)/(1-r**2)
    return t_test > t_critical
```

```python
print "Feature 1:          Feature 2:"
significance_coef = 0.01
t_critical = stats.t.ppf(1 - significance_coef, data.shape[0])
for i in range(len(columns)):
    for j in range(i + 1, len(columns)):
        test = significance_test(corr_coef[i, j], data.shape[0], t_critical)
        print "%-13s and %-13s: Correlation coefficient: %-10f; Result of significance test: %s" % \
        (columns[i], columns[j], corr_coef[i, j], str(test))
```

## Формулировка априорной гипотезы и ее проверка на значимость

## Гипотеза:
Доход университетов из ТОП-25 больше доходов высших заведений, имеющих более низкий рейтинг.

$$H_0: \mu = \mu_0,$$
$$H_1: \mu > \mu_0.$$

```python
data_rank_range = data.copy()
for k, i in enumerate(data_rank_range["world_rank"]):
    try:
        if int(i) <= 25:
            data_rank_range.loc[k, "world_rank"] = '001-025'
        if  25 < int(i) <= 100:
            data_rank_range.loc[k, "world_rank"] = '026-100'
        if 100 < int(i) <= 200:
            data_rank_range.loc[k, "world_rank"] = '101-200'
        if 200 < int(i) <= 300:
            data_rank_range.loc[k, "world_rank"]= '201-300'
        if 300 < int(i) <= 400:
            data_rank_range.loc[k, "world_rank"]= '301-400'
        if 400 < int(i) <= 500:
            data_rank_range.loc[k, "world_rank"] = '401-500'
        if 500 < int(i) <= 600:
            data_rank_range.loc[k, "world_rank"] = '501-600'
    except:
        pass
    
for k, i in enumerate(data_rank_range["world_rank"]):
    if i == '201-225' or i == '201-225' or i == '226-250' or i == '251-275' or i == '251-300' or i == '201-250' or i == '276-300':
        data_rank_range.loc[k, "world_rank"] = '201-300'
```

```python
rank_grouped = data_rank_range.groupby(data_rank_range["world_rank"]).mean()
rank_grouped
```

```python
res = stats.ttest_ind(data_rank_range[data_rank_range["world_rank"] == '001-025'].income, data_rank_range.income, nan_policy = 'omit')
print res
```

Так как p-value = 0.34874, у нас нет оснований отвергнуть нулевую гипотезу о том, что матожидание дохода университетов ТОП-25 и остальных равны.
Теперь попробуем применить эту гипотезу к вузам ТОП-25 и к "промежутку" рейтинга в отдельности

```python
res = stats.ttest_ind(data_rank_range[data_rank_range["world_rank"] == '026-100'].income, data_rank_range.income, nan_policy = 'omit')
```

```python
for k, i in enumerate(np.unique(data_rank_range["world_rank"])):
    if k == 0:
        continue
    res = stats.ttest_ind(data_rank_range[data_rank_range["world_rank"] == '001-025'].income, \
                          data_rank_range[data_rank_range["world_rank"] == i].income, nan_policy = 'omit')
    if res[1] > 0.05:
        print u'\"Промежутки рейтинга\": 001-025 и %s . Нет оснований отклонить гипотезу H0, p-value == %f' % (i, res[1])
    else:
        print u'\"Промежутки рейтинга\": 001-025 и %s . Отклоняем гипотезу H0 в пользу гипотезы H1, p-value == %f' % (i, res[1])
    
```

Доход университетов ТОП-25 не отличается существенно от доходов вузов остоальной части рейтинга, однако он в среднем выше следущей за ним ступени рейтинга от 26 до 100 места
