## Исследование списка вакансий с “The Official Website of the City of New York”
## Research_vacancies

```python
library(dplyr)
```

```python
all_data <- read.csv(file="NYC_Jobs.csv", head=TRUE, sep=",")
colnames(all_data)
data <- select(all_data, Civil.Service.Title, Salary.Range.From, Salary.Range.To, Salary.Frequency)
```

```python
head(data, 30)
```

```python
#Будем рассматривать только вакансии с ежегодной зарплатой, в силу их количественного преимущества
#Убираем повторяющиеся вакансии (т.к. некоторые постятся как внутренним сотрудникам, так и внешним (к компании) людям)

data <- filter(data, Salary.Frequency == "Annual")

data <- select(distinct(data, Civil.Service.Title), Civil.Service.Title, Salary.Range.From, Salary.Range.To)

data <- select(mutate(data, Salary = (Salary.Range.To + Salary.Range.From) / 2), Civil.Service.Title, Salary)
head(data, 10)
```

```python
hist(data$Salary, breaks = 20, xlab = "Salary distributionSalary distribution histogram",col = "lightgreen",  main = "Histogram of Salary")
```

```python
boxplot(data$Salary, main="Salary distribution boxplot", font.main=3, cex.main=1.2, ylab="Salary", font.lab=3, col="lightgreen")
```
